```python
import gensim
import matplotlib.font_manager as fm
import matplotlib
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/gdrive/Chatbot_data-master/ChatbotData .csv')
```

```python
all_data = data['Q'].values + data['A'].values
```

```python
all_text = [sentence.split(' ') for sentence in all_data]
```

```python
all_text
```

```python
model = gensim.models.Word2Vec(all_text, min_count=1, size=100, window=5)
```

```python
model.wv.vocab
```

```python
model.init_sims(replace=True)
```

```python
model.wv.most_similar('SNS')
```

```python
model.wv.most_similar('왜')
```

```python
model.wv.most_similar('다이어트')
```
